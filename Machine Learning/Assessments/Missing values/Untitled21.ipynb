```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Employee_Salaries.csv')
```

```python
df.isnull().sum()
```

```python
mode_grade = df['Grade'].mode()[0]
df['Grade'].fillna(mode_grade, inplace=True)
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
department_counts = df['Department_Name'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=department_counts.index, y=department_counts.values)
plt.title('Number of Employees in Each Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(department_counts, labels=department_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Employees Across Departments')
plt.show()
```

```python
# Line chart of average base salary over time
# Assuming 'JoiningDate' is available in the dataset
# For this example, let's create a synthetic 'JoiningDate' column
# If you have a real 'JoiningDate' column, convert it to datetime and use it
df['JoiningDate'] = pd.to_datetime('2022-01-01')
df.set_index('JoiningDate', inplace=True)
average_base_salary_per_month = df.resample('M')['Base_Salary'].mean()
plt.figure(figsize=(12, 6))
average_base_salary_per_month.plot(kind='line', marker='o')
plt.title('Average Base Salary Over Time')
plt.xlabel('Date')
plt.ylabel('Average Base Salary')
plt.show()
```

```python
# Histogram of base salaries
plt.figure(figsize=(10, 5))
sns.histplot(df['Base_Salary'], bins=20, kde=True)
plt.title('Distribution of Base Salaries')
plt.xlabel('Base Salary')
plt.ylabel('Frequency')
plt.show()
```
